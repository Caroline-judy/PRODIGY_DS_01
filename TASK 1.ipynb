```python

import pandas as pd
import zipfile

# Unzipping the provided dataset
with zipfile.ZipFile('Data Files.zip', 'r') as zip_ref:
    zip_ref.extractall('data_files')

# Listing the extracted files to understand what we have
import os
extracted_files = os.listdir('data_files')
extracted_files
```

```python
import pandas as pd

# Load the datasets
pop_data = pd.read_csv('data_files/API_SP.POP.TOTL_DS2_en_csv_v2_144171.csv', skiprows=4)
metadata_indicator = pd.read_csv('data_files/Metadata_Indicator_API_SP.POP.TOTL_DS2_en_csv_v2_144171.csv')
metadata_country = pd.read_csv('data_files/Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_144171.csv')

# Display the first few rows of each dataset
print('Population Data:')
print(pop_data.head())
print('\
Metadata Indicator:')
print(metadata_indicator.head())
print('\
Metadata Country:')
print(metadata_country.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Extract population data for the year 2020
pop_2020 = pop_data[['Country Name', '2020']].dropna()
pop_2020.columns = ['Country', 'Population_2020']

# Plotting the distribution of population in 2020
plt.figure(figsize=(12, 6))
sns.histplot(pop_2020['Population_2020'], bins=30, kde=True)
plt.title('Distribution of Population in 2020')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

# Histogram of Population Data for 2020:

• The histogram shows that the majority of countries have populations below 50 million.

• There are a few countries with very high populations, creating a long tail on the right
side of the distribution.

```python
# Extract the income group data
income_group_data = metadata_country[['Country Code', 'IncomeGroup']].dropna()

# Count the number of countries in each income group
income_group_counts = income_group_data['IncomeGroup'].value_counts()

# Plotting the distribution of income groups
plt.figure(figsize=(12, 6))
sns.barplot(x=income_group_counts.index, y=income_group_counts.values)
plt.title('Distribution of Income Groups')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.xticks(rotation=45)
plt.show()
```

This bar chart shows the number of countries in each income group.
